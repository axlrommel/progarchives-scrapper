```python
import pandas as pd
import numpy as np
import scipy
from scipy import sparse

USER_ID= 46692
USERS_TO_MATCH = 25

similarities_sparse = scipy.sparse.load_npz('./reviewer_album_stars.npz')
```

```python
def get_sparse(similarities_sparse):
    user_sparse = similarities_sparse[USER_ID]
    usr_tmp_df = pd.DataFrame(user_sparse.toarray())
    usr_tmp_series = usr_tmp_df.squeeze('rows')
    user_series = usr_tmp_series.sort_values(ascending=False).nlargest(USERS_TO_MATCH)
    return user_series
```

```python
user_series = get_sparse(similarities_sparse)
print(user_series)
```
