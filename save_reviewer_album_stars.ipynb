```python
import pandas as pd
import numpy as np
import scipy
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity

dfa = pd.read_csv('./scraper/data/reviews.txt', sep=",", names = ["reviewerId","stars","artistId","artistName","albumId","albumName"])
df = dfa[['reviewerId', 'albumId', 'stars']]
df1 = df.reviewerId.value_counts().rename_axis('unique_values').reset_index(name='counts')
df1.columns = ['reviewerId', 'count']
df2 = df1.loc[df1['count'] > 50]
df3 = df2['reviewerId'].array
prograrchives_df = df.loc[df['reviewerId'].isin(df3)]
```

```python
def get_user_item_sparse_matrix(df):
    sparse_data = sparse.csr_matrix((df.stars, (df.reviewerId, df.albumId)))
    return sparse_data
```

```python
sparse_data = get_user_item_sparse_matrix(prograrchives_df)
similarities_sparse = cosine_similarity(sparse_data, dense_output = False)
sparse.save_npz('./reviewer_album_stars', similarities_sparse)
```
